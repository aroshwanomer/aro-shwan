
# Import Libraries

Libraries used:
- pandas
- numpy
- matplotlib
- scipy


```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

print("Libraries imported successfully")
```


# Upload and Read Dataset


```

df = pd.read_csv('music_dataset.csv')

df.head()
```


# Dataset Information
This shows:
- Number of rows
- Number of columns
- Data types


```

df.info()
```


# Dataset Shape


```

print("Rows and Columns:", df.shape)
```


# Descriptive Statistics


```

df.describe()
```


# Mean


```

print("Mean Streams:", np.mean(df['Streams_Million']))
print("Mean Duration:", np.mean(df['Duration_Min']))
print("Mean Popularity:", np.mean(df['Popularity_Score']))
```


# Median


```

print("Median Streams:", np.median(df['Streams_Million']))
print("Median Popularity:", np.median(df['Popularity_Score']))
```


# Mode


```

print(df['Genre'].mode())
print(df['Mood'].mode())
```


# Standard Deviation


```

print("STD Streams:", np.std(df['Streams_Million']))
print("STD Duration:", np.std(df['Duration_Min']))
print("STD Popularity:", np.std(df['Popularity_Score']))
```


# Variance


```

print("Variance Popularity:", np.var(df['Popularity_Score']))
```


# Quartiles


```

Q1 = df['Popularity_Score'].quantile(0.25)
Q2 = df['Popularity_Score'].quantile(0.50)
Q3 = df['Popularity_Score'].quantile(0.75)

print("Q1 =", Q1)
print("Q2 =", Q2)
print("Q3 =", Q3)
```


# Correlation


```

numeric_df = df[['Streams_Million', 'Duration_Min', 'Popularity_Score']]

numeric_df.corr()
```


# Genre Count


```

df['Genre'].value_counts()
```


# Bar Chart


```

genre_counts = df['Genre'].value_counts()

genre_counts.plot(kind='bar')

plt.xlabel("Genre")
plt.ylabel("Count")
plt.title("Music Genre Distribution")

plt.show()
```


# Histogram


```

plt.hist(df['Popularity_Score'], bins=10)

plt.xlabel("Popularity Score")
plt.ylabel("Frequency")
plt.title("Popularity Histogram")

plt.show()
```


# Pie Chart


```

df['Mood'].value_counts().plot(kind='pie', autopct='%1.1f%%')

plt.title("Mood Distribution")

plt.show()
```


# Box Plot


```

plt.boxplot(df['Popularity_Score'])

plt.title("Boxplot of Popularity Score")

plt.show()
```
